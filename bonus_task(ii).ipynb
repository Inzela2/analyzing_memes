```python
import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow


mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=1, min_detection_confidence=0.5)

def process_image(image_path):

    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


    results = hands.process(image_rgb)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:

            # mp.solutions.drawing_utils.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)


            tip_y = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].y
            mcp_y = hand_landmarks.landmark[mp_hands.HandLandmark.MIDDLE_FINGER_MCP].y
            if tip_y < mcp_y:
                print("Middle finger might be raised, potentially toxic meme")
            else:
                print("Middle finger is not raised, potentially not a toxic meme")
    else:
        print("No hands detected.")



    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


process_image('/content/middle_finger.png')
```
